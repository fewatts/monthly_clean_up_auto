```python
import json
import pandas as pd
import lib.mass_update as mu

# --- FILE PATH SETUP ---
mosaiq_hierarchy = r'C:\Users\falve11887\OneDrive - Elekta\Documents\Monthly Clean up automations\products_hierarchy\MOSAIQ.json'

# --- LOAD HIERARCHY DATA ---
# Load the complete JSON hierarchy structure directly without modifications
with open(mosaiq_hierarchy, 'r', encoding='utf-8') as file:
    mosaiq_hierarchy_dict = json.load(file)

# --- RECURSIVE EXTRACTION FUNCTION ---
def extract_all_product_codes(nodes):
    """
    Recursively traverses the entire JSON structure to collect 
    all unique 'Product Code' values found across all tree branches.
    """
    extracted_codes = []
    # Ensure the input node/nodes structure is treated as a list
    node_list = nodes if isinstance(nodes, list) else [nodes]
    
    for node in node_list:
        if isinstance(node, dict):
            if "Product Code" in node:
                extracted_codes.append(node["Product Code"])
            # If child nodes exist, dive deeper down the tree branches
            if "Children" in node:
                extracted_codes.extend(extract_all_product_codes(node["Children"]))
                
    # Use set() to eliminate any duplicated codes from the final list
    return list(set(extracted_codes))

# --- EXECUTE EXTRACTION ---
# Get all unique raw codes directly from the complete hierarchy dictionary
final_extracted_codes = extract_all_product_codes(mosaiq_hierarchy_dict)

# --- CHUNK PACKING ALGORITHM ---
# Set the maximum character length threshold allowed per report block
CHARACTER_LIMIT = 1240 

final_report_blocks = []
current_working_block = []

for code in final_extracted_codes:
    # Calculate what the total character length would be if this code was appended
    potential_length = len(",".join(current_working_block + [code]))
    
    if potential_length <= CHARACTER_LIMIT:
        current_working_block.append(code)
    else:
        # If adding the code breaches the threshold, pack the current block and reset
        final_report_blocks.append(",".join(current_working_block))
        current_working_block = [code]

# Append the final remaining codes block if it contains data
if current_working_block:
    final_report_blocks.append(",".join(current_working_block))

# --- PRINT REPORT BLOCKS ---
# Output each individual text block with its verified character metadata
for index, block in enumerate(final_report_blocks, 1):
    print(f"--- BLOCK {index} ({len(block)} characters) ---")
    print(block)
    print("\n")
```
